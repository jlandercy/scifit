```python
import numpy as np
from scipy import fft
import matplotlib.pyplot as plt
```

```python
f = 10000  # Hz
T = 1.5     # s
t = np.linspace(0, T, int(f*T))
```

```python
x = 10*np.sin(2*np.pi*50*t) + 5*np.sin(2*np.pi*100*t) + np.sin(2*np.pi*150*t)
```

```python
X = fft.fft(x)
```

```python
def bode(X, f=1.):
    
    N = X.size // 2
    freq = fft.fftfreq(X.size, 1/f)

    fig, axes = plt.subplots(2, 1, sharex=True, gridspec_kw={"hspace": 0.05})
    axes[0].loglog(freq[0:N], np.abs(X[0:N]))
    axes[1].semilogx(freq[0:N], np.angle(X[0:N]))
    
    for i in range(2):
        axes[i].grid()
        
    axes[0].set_title("Bode plot")
    axes[0].set_ylabel("abs(X)")
    axes[1].set_ylabel("arg(X)")
    axes[1].set_xlabel("Frequency")
    
    return axe
```

```python
axe = bode(X, f=f)
```
