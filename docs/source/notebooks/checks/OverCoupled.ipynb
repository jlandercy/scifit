```python
import numpy as np
from scifit.solvers import FitSolver1D
```

```python
class Model(FitSolver1D):
    @staticmethod
    def model(x, a, b):
        return (x[:,0]/x[:,0])*(a * b)
```

```python
model = Model()
```

```python
data = model.synthetic_dataset(parameters=(2, 5), sigma=0.1)
```

```python
sol = model.fit(data)
```

```python
axe = model.plot_fit()
```

```python
axe = model.plot_loss(factor=10, levels=15, log_loss=False)
```
