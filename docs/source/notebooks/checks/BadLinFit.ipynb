```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scifit.solvers import FitSolver1D
```

```python
class Model(FitSolver1D):
    @staticmethod
    def model(x, a, b):
        return a**2 * x[:, 0] + b
```

```python
domains = pd.DataFrame({"min": [-2, -2], "max": [2, 0]}).T
```

```python
model = Model()
```

```python
data = model.synthetic_dataset(xmin=-1, xmax=+1, sigma=0.1, parameters=(-1, -1))
```

```python
sol1 = model.fit(data)
```

```python
axe = model.plot_fit()
axe.figure.savefig("figures/NonLinear/BadLineFit.pdf")
```

```python
axe = model.plot_loss()
axe.figure.savefig("figures/NonLinear/BadLineLoss.pdf")
```

```python
axe = model.plot_loss(domains=domains, levels=20)
axe.figure.savefig("figures/NonLinear/BadLineLossZoom.pdf")
```
