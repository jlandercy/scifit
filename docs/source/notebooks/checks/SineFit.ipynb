```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scifit.solvers import FitSolver1D
```

```python
class Sine(FitSolver1D):
    @staticmethod
    def model(x, a, b):
        return np.sin(a * x[:, 0] + b)
```

```python
domains = pd.DataFrame({"min": [-2*np.pi, -2*np.pi], "max": [+2*np.pi, +2*np.pi]}).T
```

```python
model = Sine()
```

```python
data = model.synthetic_dataset(xmin=-10, xmax=+10, sigma=0.1, parameters=(-1, -1))
```

```python
sol1 = model.fit(data)
```

```python
axe = model.plot_fit()
axe.figure.savefig("figures/NonLinear/SineFit.pdf")
```

```python
axe = model.plot_loss()
axe.figure.savefig("figures/NonLinear/SineLoss.pdf")
```

```python
axe = model.plot_loss(domains=domains)
axe.figure.savefig("figures/NonLinear/SineLossZoom.pdf")
```

```python
sol2 = model.fit(data, p0=[0, -3])
```

```python
axe = model.plot_fit()
axe.figure.savefig("figures/NonLinear/SineFitBad.pdf")
```

```python
axe = model.plot_loss(domains=pd.DataFrame({"min": [-1,-5], "max": [1, -2]}).T)
axe.figure.savefig("figures/NonLinear/SineLossBad.pdf")
```
