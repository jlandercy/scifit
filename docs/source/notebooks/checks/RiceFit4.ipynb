```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import special, stats
```

```python
def model(x, b):
    return x*np.exp(-(x**2 + b**2)/2)*special.i0(x*b)
```

```python
def model2(b, s, x=1.):
    return (x/s)*np.exp(-((x/s)**2 + b**2)/2)*special.i0((x/s)*b)
```

```python
def model3(b, s, x=1.):
    return stats.rice.pdf(x, b=b, loc=0., scale=s)
```

```python
x = np.linspace(0, 20, 200)
```

```python
fig, axe = plt.subplots()
axe.plot(x, model(x, 5))
```

```python
xlin = np.linspace(0, 10, 200)
ylin = np.linspace(0, 2, 200)
X, Y = np.meshgrid(xlin, ylin)
```

```python
Z = model(X, Y)
```

```python
Z2 = model2(X, Y, x=1) + model2(X, Y, x=2) + model2(X, Y, x=3)
```

```python
fig, axe = plt.subplots()
axe.contour(X, Y, Z, 30, cmap="jet")
axe.grid()
```

```python
fig, axe = plt.subplots()
axe.contour(X, Y, Z2, 30, cmap="jet")
axe.grid()
```

```python
Z2 = (0 - model3(X, Y, x=1))**2 + (0 - model3(X, Y, x=2))**2
```

```python
fig, axe = plt.subplots()
axe.contour(X, Y, Z2, 30, cmap="jet")
axe.grid()
```
