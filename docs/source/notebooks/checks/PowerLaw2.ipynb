```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize
from scifit.solvers import FitSolver1D
```

```python
class Model(FitSolver1D):
    @staticmethod
    def model(x, alpha):
        return (alpha - 1.) * np.power(x[:, 0], -alpha)
```

```python
model = Model()
```

```python
data = model.synthetic_dataset(xmin=1, xmax=50, parameters=(2.5,), sigma=0.015, scale_mode="rel")
```

```python
sol = model.fit(data, p0=(3.,))
```

```python
model.plot_fit(log_x=True, log_y=True)
```

```python
axe = model.plot_loss(log_loss=False, domains=pd.DataFrame({"min": [0.90], "max": [3.]}).T, resolution=500)
axe.set_yscale("log")
```
