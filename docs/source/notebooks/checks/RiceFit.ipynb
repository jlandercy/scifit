```python
import numpy as np
import pandas as pd
from scifit.solvers import FitSolver1D
from scipy import special
```

```python
class Model(FitSolver1D):
    @staticmethod
    def model(x, nu, sigma):
        return (x[:,0]/sigma**2)*np.exp(-0.5*(x[:,0]**2 + nu**2)/sigma**2)*special.i0(x[:,0]*nu/sigma**2)
```

```python
model = Model()
```

```python
data = model.synthetic_dataset(xmin=0, xmax=50, parameters=(5, 7), sigma=0.0001)
```

```python
sol = model.fit(data)
```

```python
model.plot_fit()
```

```python
domains = pd.DataFrame({"min": [-10, -10], "max": [10, 10]}).T
model.plot_loss(domains=domains, log_loss=True)
```

```python
domains = pd.DataFrame({"min": [0, 0], "max": [10, 10]}).T
model.plot_loss(domains=domains, log_loss=True)
```
