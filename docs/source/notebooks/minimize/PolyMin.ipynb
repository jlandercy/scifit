```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def model(z):
    x, y = z
    return 0.5 * (x**4 - 16*x**2 + 5*x + y**4 - 16*y**2 + 5*y)
```

```python
xlin = np.linspace(-4, 4, 200)
ylin = np.linspace(-4, 4, 200)
X, Y = np.meshgrid(xlin, ylin)
```

```python
Z = model([X, Y])
```

```python
fig, axe = plt.subplots(subplot_kw={"projection": "3d"})
surface = axe.plot_surface(X, Y, Z, cmap="jet", alpha=0.75)
```

```python
fig, axe = plt.subplots()
contours = axe.contour(X, Y, Z, 10, cmap="jet")
axe.clabel(contours, contours.levels, inline=True)
axe.set_aspect("equal")
axe.grid()
```

```python
optimize.minimize(model, x0=[1, 1])
```

```python
optimize.minimize(model, x0=[0, 0])
```

```python
optimize.basinhopping(model, niter=100, x0=[1, 1])
```

```python
optimize.basinhopping(model, niter=1000, x0=[1, 1])
```

```python
optimize.basinhopping(model, stepsize=5, x0=[1, 1])
```

```python
optimize.brute(model, ((-4, 4), (-4, 4)), finish=optimize.minimize)
```

```python
optimize.differential_evolution(model, ((-4, 4), (-4, 4)))
```

```python
optimize.shgo(model, ((-4, 4), (-4, 4)))
```

```python
optimize.dual_annealing(model, ((-4, 4), (-4, 4)))
```

```python
optimize.direct(model, ((-4, 4), (-4, 4)))
```
